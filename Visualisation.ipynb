**Visualization in pandas**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\user\Documents\Ice Cream Ratings.csv")
df = df.set_index('Date')
df
```

```python
print(plt.style.available)
plt.style.use('tableau-colorblind10')
```

```python
df.plot(kind ='line', title = 'Icecream Ratings', xlabel = 'Daily Rating', ylabel = 'Scores', figsize = (7,5))
```

```python
df.plot(kind = 'bar', xlabel = 'Daily Rating', ylabel = 'Scores',  stacked = True, figsize = (7,5) )
```

```python
df.plot.scatter(x='Texture Rating', y= 'Overall Rating', s = 350, c ='Grey', figsize = (7,5))
```

```python
df.plot.hist(bins =20, figsize = (7,5))
```

```python
df.boxplot(figsize = (7,5))
```

```python
df.plot.area(figsize = (7,5))
```

```python
df.plot.pie(y = 'Flavor Rating', figsize = (9,8))
```
